# Installera Python och Miniforge
Det finns många sätt att installera Python och en pakethanterare på sin egen, eller Chalmers datorer. Här presenterar vi en variant. Miniforge är en terminalbaserad pakethanterare för Python-paket liknande conda, men med fördelen att den är mycket (!!!) snabbare.

## OS X (Mac) och Linux
- Gå till sidan https://github.com/conda-forge/miniforge, och skrolla ner till rubriken "Download".
- Ladda ner programmet som matchar ditt operativsystem:
    - Mac OS X (Intel processor): Miniforge3-MacOSX-x86_64
    - Mac OS X (M* processor): Miniforge3-MacOSX-arm64
    - Linux (ex. Chalmersdator): Miniforge3-Linux-x86_64
- Öppna en terminal och navigera till mappen som heter "Downloads": 
    ```
    cd ~/Downloads/
    ```
    Kolla så att den nya filen finns där.
- Kör scriptet genom att i terminalen skriva (...-Linux-x86_64 ska matcha den filen du laddade ner):
    ```
    bash Miniforge3-Linux-x86_64
    ```
    Tryck på Enter och följ instruktionerna. Här kan du svara "yes" på allt.
- När installationen är klar ska du starta om din terminal. Om allt har gått rätt bör din terminal se ut något såhär
    ```
    $ (base) ditt_namn@....
    ```
    (base) indikerar här att det är Python-miljön base som är aktiv. Kommandot "which" följt av python bör också returnera något som liknar detta:
    ```
    which python
    /home/erik/miniforge3/bin/python
    ```

När du kommit såhär långt kan du gå vidare till "Installera Python-paket med mamba"

## Windows
- Gå till sidan https://github.com/conda-forge/miniforge, och skrolla ner till rubriken "Download".
- Ladda ner programmet för Windows: Miniforge3-Windows-x86_64
- Kör programmet och följ instruktionerna. Du kan behöva trycka på "More info" för att få fram "Run anyway".
    - Installera för "Just me"
    - Se till att valet att skapa en genväg i startmenyn är i-klickat.
- Testa att starta en "Miniforge Prompt" via startmenyn.

## Installera Python-paket med mamba

Här följer instruktioner för att installera de paket ni kommer behöva för den här kursen i pythonmiljön "base". I framtiden kommer ni antagligen vilja skapa flera olika miljöer beroende på projekt, men det är lite överkurs just nu.

- Öppna en terminal (Mac OSX/Linux) eller Miniforge Prompt (Windows)
- Skriv följande i terminalen
    ```
    mamba install numpy matplotlib jupyterlab
    ```
    och svara "yes" när det efterfrågas.
- När detta är klart har ni alla paket som behövs.
- Skulle ni upptäcka att ni saknar ett paket kan ni alltid installera det senare. Till exempel:
    ```
    mamba install spyder
    ```

För att arbeta i t.ex. Jupyter Lab i fortsättningen öppnar ni en terminal eller Miniforge Prompt och skriver 
```
jupyter lab
```
Detta kommer öppna en ny flik i er webbläsare där ni kan jobba med notebooks.
